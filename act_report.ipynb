## Report: act_report

Based on the tweets by WeRateDogs Twitter Archive tweeted from 2015-11-15 to 2017-08-01, the following findings we made.

#### 1. Most of the tweets sent on to that account we posted using an iPhone.

Tweets posted using an iPhone summed up to 2125 tweets while the other sources used were:
- Vine - Make a Scene which was used to post 91 tweets during that duration.
- Twitter Web Client used to post 31 tweets in that duration of time and,
- TweetDeck used to post only 11 tweets in that period.

Below is a horizontal bar graph to show the distribution of the number of tweets posted by the various means.


#### 2. Most of the breed were predicted using the first image.

The dog breeds predictions were made by running every image in the WeRateDogs Twitter archive through a neural network that can classify breeds of dogs. Up to four pictures were used to predict the dog's breed but a majority of the dog's breed were predicted using the first image as show in the horizontal bar graph distribution below.


#### 3. Most of the dogs were puppers.

Although some of the dogs stages were not identified, for those that were, a majority of them were puppers. That can be depicted from the graph shown herein.
- Puppers were the majority, 235 out of the 362 dogs whose dog stages were given.
- Doggos were 91.
- Puppos were 27 and,
- Floofers were 9.
